# Data Visualization Gallery

Images of all current data visualizations I have created. The below script scans all folders to print out all saved images.

## Import Packages

```python
import os

from IPython.display import Image
```

## Create a function and set paths
Function display images when a path and filename is set.

```python
## Create a function to display images
def slide(img):
    display( Image(img, height = 200) )
    
    
## set cwd
path = os.getcwd() 


## image folder name
imageFolder="\\images"

folders = os.listdir()
```

## Print Images
The function below loops through all folders in data projects. Searches each folder for an images folder. If an images folder exists and at least one file is available, print the file(s).

```python
for i in folders:
    ## Set the path to the CWD, add a variable, then search for the images folder
    check=path + "\\" + i + "\\" + imageFolder + "\\"
    ## Loop through all folders in data projects
    if os.path.isdir(check):
        ## Check to see if files exist
        listFiles = os.listdir(check)
        ## If files exist, use user function to dispaly all files
        for f in listFiles:
            print(f)
            slide(check + f)
```
