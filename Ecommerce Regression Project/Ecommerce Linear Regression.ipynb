```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df_raw = pd.read_csv(r'https://github.com/pestyld/data/raw/main/Ecommerce%20Customers')
df_raw.head()
```

```python
df_raw.info()
```

```python
df_raw.isna().sum()
```

```python
df_raw.Email.str.split('@').str[1]
```

## Prepare data
- Rename columns
- Parse out email domain
- Parse out State and Zip

```python
df = (
    df_raw
    .rename(columns = lambda colname : colname.replace('.',' ').replace(' ',''))
    .assign(
        EmailDomain = df_raw.Email.str.split('@').str[1],
        Address = df_raw.Address.replace('\n',' ', regex=True),
        StateZip = lambda _df: _df.Address.str.split(',').str[1].str.strip(),
        State = lambda _df: _df.StateZip.str.split(' ').str[0],
        Zip = lambda _df: _df.StateZip.str.split(' ').str[1]
    )
    .drop('StateZip', axis='columns')
)

df.head()
```

## Explore data

View distribution of the target (YearlyAmountSpent)

```python
sns.histplot(data = df, x = 'YearlyAmountSpent');
```

Compare YearlyAmount Spent with TimeonApp and TimeonWebsite.

```python
viewCorrColumns = ['TimeonApp', 'TimeonWebsite']
target = 'YearlyAmountSpent'
for col in viewCorrColumns:
    sns.jointplot(data = df, x = col, y = target, kind="reg")
```

View correlation with all features (minus categoricals)

```python
sns.pairplot(df[['AvgSessionLength', 'TimeonApp',
       'TimeonWebsite', 'LengthofMembership', 'YearlyAmountSpent']])
```

Summary: Results of this show that LengthofMembership seems to be highly correlated with YearlyAmountSpent.

```python
sns.lmplot(data = df, x = 'LengthofMembership', y = 'YearlyAmountSpent')
```

## Create the model

```python
df.head()
```

### Training and testing data

```python
y = df['YearlyAmountSpent']
X = df.loc[:,['AvgSessionLength', 'TimeonApp', 'TimeonWebsite', 'LengthofMembership']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

### Create the model

```python
lm = LinearRegression()

lm.fit(X_train,y_train)

pd.DataFrame({'Values':np.append(lm.coef_, lm.intercept_)}, 
             index = X.columns.insert(len(X.columns),'Intercept'))
```

### Predict values

```python
predictions = lm.predict(X_test)
```

```python
sns.regplot(x = y_test, y = predictions)
plt.xlabel('Real Values')
plt.ylabel('Prediction');
```

### Evaluating model

```python
print('MAE :', metrics.mean_absolute_error(y_test, predictions))
print('MSE :', metrics.mean_squared_error(y_test, predictions))
print('RMSE :', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
metrics.r2_score(y_test, predictions)
```

```python
metrics.explained_variance_score(y_test, predictions)
```

### Residuals

```python
residuals = y_test - predictions
```

```python
sns.histplot(residuals, bins = 50)
plt.title('Residuals');
```

## Conclusion

Holding all other features fixed:
- A one unit increase in **AvgSessionLength** is associated with an increase of **\$26** spent.
- A one unit increase in **TimeonApp** is associated with an increase of **\$39** spent.
- A one unit increase in **TimeonWebsite** is associated with an increase of **\$.19** spent.
