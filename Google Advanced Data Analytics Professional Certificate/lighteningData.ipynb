```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn.objects as so
```

```python
df_raw = pd.read_csv('data/lightning_strikes.csv', nrows = 300000)
```

```python
df_raw.head()
```

```python
df_raw.dtypes
```

```python
df = (
    df_raw
    .assign(
        DATE = pd.to_datetime(df_raw['DATE'].astype(str),format='%Y%m%d'),
        year = lambda _df: _df.DATE.dt.year,
        year_str = lambda _df: _df.DATE.dt.strftime('%Y').astype('category'),
        month = lambda _df: _df.DATE.dt.month,
        year_month_str = lambda _df: _df.DATE.dt.strftime('%Y-%m').astype('category'),
        quarter = lambda _df: _df.DATE.dt.quarter,
        quarter_period = lambda _df: _df.DATE.dt.to_period('Q'),
        quarter_str = lambda _df: _df.DATE.dt.to_period('Q').dt.strftime('%Y-Q%q').astype('category'),
        weeks = lambda _df : _df.DATE.dt.isocalendar().week,
        weeks_str = lambda _df: _df.DATE.dt.strftime('%Y-W%V'),
        weekday = lambda _df: _df.DATE.dt.day_name().astype('category'),
        center_point_geom = lambda _df: '(' + _df.CENTERLON.astype('str').str.cat(_df.CENTERLAT.astype('str'), sep=', ')+')'
           )
    .drop_duplicates()
    .rename(columns = lambda colName: colName.lower())
)
```

```python
lightening_by_month = (df[['year_str','month', 'total_count']]
                     .groupby(['year_str','month'])
                     .sum()
                     .reset_index()
                     .rename(columns={'total_count':'number_of_strikes'})
                    )
lightening_by_month
```

```python
lightening_by_month = (df[['year_str','month', 'total_count']]
                     .groupby(['year_str','month'])
                     .agg(sum)
                     .reset_index()
                     .rename(columns={'total_count':'number_of_strikes'})
                    )
lightening_by_month
```

```python
lightening_by_month = (df[['year_str','month', 'total_count']]
                       .groupby(['year_str','month'])
                     .  agg(number_of_strikes = pd.NamedAgg(column='total_count', aggfunc=sum))
                       .reset_index()
                    )
lightening_by_month
```

```python
lightening_by_year = (df[['year_str','total_count']]
                     .groupby('year_str')
                     .sum()
                     .reset_index()
                    )
lightening_by_year
```

```python
pct_df = \
  lightening_by_month \
 .merge(lightening_by_year) \
 .assign(pct = lambda _df: _df.number_of_strikes/_df.total_count)

pct_df
```

```python
ax = sns.barplot(data = pct_df, 
                 x = 'month', 
                 y='pct', 
                 hue='year_str')

ax.set_xlabel('Month')
ax.set_ylabel('% of lightning strikes')
ax.set_title('% of lightning strikes each Month');
```
