```python
def prepare_diamond_data():
    
    ##
    ## Read in the raw data
    ##
    
    path = r'https://github.com/pestyld/data/raw/main/diamonds.csv'
    
    # use index_col = 0 to use the first column as the index.
    raw_data = pd.read_csv(path, index_col = [0])
    
    ## Numeric columns: ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
    numCols = df.select_dtypes('number').columns.to_list()

    ## Categorical columns
    catCols = ['cut','color','clarity']
    
    
    ##
    ## Create a new dataframe
    ##
    final_data = (
        raw_data
        .query('~(x == 0 and y == 0 and z ==0)')                ## Filter out rows where x = 0, y = 0 and z = 0 (weird when using 3 ands in query method.
        .assign(
            x = lambda df: df.x.replace(0, df.x.median()),      ## Replace 0 values with the median for each column (x/z)
            z = lambda df: df.z.replace(0, df.z.median())
        )
        .query('z < 30 and y < 30')                             ## Remove outliers from z and y
        .reset_index(drop = True)                               ## Reset the index for an accumulating value
    )
    
    ## Delete the raw datafram
    del raw_data
    
    ## One hot encoding to create dummy columns
    for i in catCols:
        final_data = pd.concat([final_data, pd.get_dummies(final_data[i], drop_first=True)], axis=1)
    
    ## Return final DataFrame
    return final_data

newdf = prepare_diamond_data()

newdf.head()
```
