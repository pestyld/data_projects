```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

## Create Data

```python
x = np.linspace(1,5,11)
```

```python
y = x **2
```

```python
x
```

```python
y
```

## Functional Method

```python
plt.plot(x,y,label='My Line Label')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('This is my title')
plt.legend()
plt.show() #<--need this outside jupyter notebook
```

#### Creating sub plots

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g')
plt.subplot(1,2,2)
plt.plot(y,x,'r')
```

## Object Oriented Method

#### Completing the folloing to create a plot:
1. Creating the initial blank figure
2. Add percent of the blank canvas to use with the add_axes. Use percentages, [left,bottom,width,height]
3. Plot on the axes

```python
fig = plt.figure()
axes=fig.add_axes([.1,.1,.8,.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title("This is a title")
```

#### Creating multiple plots

```python
fig = plt.figure()
axes1=fig.add_axes([.1,.1,.8,.8])
axes2=fig.add_axes([.2,.5,.4,.3])
```

```python
fig = plt.figure()
axes1=fig.add_axes([.1,.1,.8,.8])
axes2=fig.add_axes([.1,.6,.4,.3])
axes1.plot(x,y,'o')
axes2.plot(y,x,'g')
```

#### Create subplots

- Creates an array for axes

```python
fig,axes = plt.subplots(2,2)
type(axes)
```

#### Plot on the subplots

```python
fig,axes = plt.subplots(2,2)
axes[0,0].plot(x,y)
axes[0,0].set_title("Axis 0,0")

axes[0,1].plot(y,x,'g')
axes[0,0].set_title("Axis 0,1")

axes[1,0].plot(y,x,'r')
axes[0,0].set_title("Axis 1,0")

axes[1,1].plot(x,y,'o')
axes[0,0].set_title("Axis 1,1")
```
