```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

## Create Data

```python
x = np.linspace(1,5,11)
```

```python
y = x **2
```

```python
x
```

```python
y
```

## Figure Size and DPI

#### Figsize is in inches. Set as a tuple. Can add DPI (pixels)

```python
fig = plt.figure(figsize=(3,2),dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2),dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,4),dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

#### Works the same with subplots

```python
fig,axes = plt.subplots(1,2,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```
