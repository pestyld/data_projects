```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

## Create Data

```python
x = np.linspace(1,5,11)
```

```python
y = x **2
```

```python
x
```

```python
y
```

## Basic Plot Appearance Options

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',linewidth=3,markersize=10,marker='o',linestyle='--',alpha=.2)
ax.plot(y,x,color='green',linewidth=5,markersize=15,marker='o',
        markerfacecolor='yellow',markeredgewidth=3, markeredgecolor='b',linestyle='-.',alpha=.6)
```

## X and Y Axes Lower and Upper Bound

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlim([0,10])
ax.set_ylim([0,10])
```
