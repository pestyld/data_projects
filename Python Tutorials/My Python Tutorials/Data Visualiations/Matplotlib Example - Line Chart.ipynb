## 1. Plot one line

```python
plt.plot(ages_x,py_dev_y, marker='o',markersize=4, label="Python")
```

####  a.Enhance Visualization

```python
plt.plot(ages_x,py_dev_y, marker='o',markersize=4, label="Python")

plt.title("Median Salaries (USD) by Age for Python Developers")
plt.ylabel("Median Salary")
plt.xlabel("Age")
plt.legend()
plt.show()
```

## 2. Plot two lines. Add formatting options

```python
plt.plot(ages_x,py_dev_y, marker='o',markersize=4, color='k', label="Python")
plt.plot(ages_x,js_dev_y, marker='o',markersize=4, linestyle='--', label="JavaScript")

#enhance graph
plt.title("Median Salaries (USD) by Age for Python & JavaScript Developers")
plt.ylabel("Median Salary")
plt.xlabel("Age")
plt.legend()

plt.show()
```

#### a. Enlarge visual

```python
#plot two line. Add formatting options
plt.figure(figsize=(10,8))
plt.plot(ages_x,py_dev_y, marker='o',markersize=4, color='k', label="Python")
plt.plot(ages_x,js_dev_y, marker='o',markersize=4, linestyle='--', label="JavaScript")

#enhance graph
plt.title("Median Salaries (USD) by Age for Python & JavaScript Developers")
plt.ylabel("Median Salary")
plt.xlabel("Age")
plt.legend()

plt.show()  
```

## 3. Plot three lines

```python
plt.figure(figsize=(10,8))
plt.plot(ages_x,py_dev_y, marker='o',markersize=4, color='k',linewidth=3, label="Python")
plt.plot(ages_x,js_dev_y, marker='o',markersize=4,linewidth=3, label="JavaScript")
plt.plot(ages_x,dev_y, linestyle='--', color='r', label="All Devs")

#enhance graph
plt.title("Median Salaries (USD) by Age for Python & JavaScript Developers")
plt.ylabel("Median Salary")
plt.xlabel("Age")
plt.legend()

plt.show()  
```

#### a. Enhance visual with grid and tight layout

```python
#plot 3 line. Add formatting options. Add grid + tight layout
plt.figure(figsize=(10,8))
plt.plot(ages_x,py_dev_y, marker='o',markersize=4, color='k',linewidth=3, label="Python")
plt.plot(ages_x,js_dev_y, marker='o',markersize=4,linewidth=3, label="JavaScript")
plt.plot(ages_x,dev_y, linestyle='--', color='r', label="All Devs")
#enhance graph
plt.title("Median Salaries (USD) by Age for Python & JavaScript Developers")
plt.ylabel("Median Salary")
plt.xlabel("Age")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()  

```

#### c. Apply the fivethirtyeight style to the visual. Still using specified options in plt.plot

```python
#plot 3 line. Add formatting options. Add style
plt.figure(figsize=(10,8))
plt.style.use('fivethirtyeight')
plt.plot(ages_x,py_dev_y, marker='o',markersize=4, color='k',linewidth=3, label="Python")
plt.plot(ages_x,js_dev_y, marker='o',markersize=4,linewidth=3, label="JavaScript")
plt.plot(ages_x,dev_y, linestyle='--', color='r', label="All Devs")

#enhance graph
plt.title("Median Salaries (USD) by Age for Python & JavaScript Developers")
plt.ylabel("Median Salary")
plt.xlabel("Age")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()  
```
