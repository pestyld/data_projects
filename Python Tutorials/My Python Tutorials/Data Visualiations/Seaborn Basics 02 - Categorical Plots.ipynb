```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

## Create Data

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

## Barplot Plot

#### By default y axis finds the mean

```python
sns.barplot(x='sex',y='size',data=df)
```

#### Find the std for each group

```python
sns.barplot(x='sex',y='size',data=df,estimator=np.std)
```

## Count Plot

```python
sns.countplot(x='sex',data=df)
```

## Box Plots

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df, hue='smoker')
```

## Violin Plots

```python
sns.violinplot(x='day',y='total_bill',data=df)
```

```python
sns.violinplot(x='day',y='total_bill',data=df,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=df,hue='smoker',split=True)
```

## Strip Plot

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
sns.stripplot(x='day',y='total_bill',data=df, hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=df, hue='sex', dodge=True)
```

## Swarm Plot

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df, color='black')
sns.violinplot(x='day',y='total_bill',data=df)
```

## Cat Plot

```python
sns.catplot(x='day',y='total_bill',data=df)
```
