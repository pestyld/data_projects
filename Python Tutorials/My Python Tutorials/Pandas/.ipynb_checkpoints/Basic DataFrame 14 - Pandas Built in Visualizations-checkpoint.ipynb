```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df1=pd.read_csv('df1',index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

## Histogram

```python
df1.hist(column='A',bins=30)
```

```python
df1.A.plot(kind='hist',bins=30)
```

```python
df1.A.plot.hist(bins=30)
```

## Area Plot

```python
df2.plot.area(alpha=.6)
```

## Bar Plot

##### Need a categorical index

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

## Horizontal Bar

```python
df2.plot.barh(stacked=True)
```

## Box Plot

```python
df2.head()
```

```python
df2.plot.box()
```

## Density or KDB Plot

```python
df2.plot.density()
```

```python
df2.plot.kde()
```

## LIne Plot

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

#### Make size of points a column

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*100,)
```

## Hexbin Plot

```python
df=pd.DataFrame(np.random.randn(1000,2),columns=['A','B'])
```

```python
df.plot.hexbin(x='A',y='B',gridsize=20, cmap='coolwarm')
```
