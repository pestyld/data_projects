## Creating and deleting columns
There are several columns that contain data on the number of Facebook likes. Let's add up all actor and director Facebook like columns and assign them to the total_likes column. We can do this in a couple of ways.

### Create a simple column called has_seen with a value of 0.

```python
def clean_columns(val):
    return val.strip().lower().replace(' ','_')

(movies
 .set_index('movie_title')
 .rename(index = index_rename, columns=clean_columns)
 .assign(has_seen = 0)
 .head()
)
```

### Add a lambda function within the assign statement.

```python
def clean_columns(val):
    return val.strip().lower().replace(' ','_')

fb_likes_sum = ("actor_1_facebook_likes","actor_2_facebook_likes","actor_3_facebook_likes","director_facebook_likes")

(movies
 .set_index('movie_title')
 .rename(columns=clean_columns)
 .assign(total_likes = lambda _df: _df.loc[:,fb_likes_sum].sum(axis = 'columns'))
)
```

### Create a function and use it in the assign method

```python
def clean_columns(val):
    return val.strip().lower().replace(' ','_')

def sum_likes(_df):
    fb_likes_sum = ("actor_1_facebook_likes","actor_2_facebook_likes","actor_3_facebook_likes","director_facebook_likes")
    return _df.loc[:,fb_likes_sum].sum(axis = 'columns')

(movies
 .set_index('movie_title')
 .rename(columns=clean_columns)
 .assign(total_likes = sum_likes)
)
```

### Use a fancy function to find specific columns

```python
def clean_columns(val):
    return val.strip().lower().replace(' ','_')

## Find all columns with likes and (actor or director)
def sum_likes(_df):
    sum_cols = []
    for c in _df.columns:
        if "likes" in c and ("actor" in c or "director" in c):
            sum_cols.append(c)
    return _df.loc[:,sum_cols].sum(axis = "columns")



(movies
 .set_index('movie_title')
 .rename(columns=clean_columns)
 .assign(total_likes = sum_likes)
 .head()
)
```

### Create a new column that is a boolean value
There is another column in the dataset named cast_total_facebook_likes. It would be interesting to see what percentage of this column comes from our newly created column, total_likes. Before we create our percentage column, let's do some basic data validation. We will ensure that cast_total_facebook_likes is greater than or equal to total_likes:

#### Using a lambda function

```python
def clean_columns(val):
    return val.strip().lower().replace(' ','_')

def sum_likes(_df):
    fb_likes_sum = ("actor_1_facebook_likes","actor_2_facebook_likes","actor_3_facebook_likes","director_facebook_likes")
    return _df.loc[:,fb_likes_sum].sum(axis = 'columns')

(movies
 .set_index('movie_title')
 .rename(columns=clean_columns)
 .assign(total_likes = sum_likes,
         is_cast_likes_more = lambda _df: _df.cast_total_facebook_likes >= _df.total_likes)
 .head()
)
```

#### Using a defined function

```python
def clean_columns(val):
    return val.strip().lower().replace(' ','_')

def sum_likes(_df):
    fb_likes_sum = ("actor_1_facebook_likes","actor_2_facebook_likes","actor_3_facebook_likes","director_facebook_likes")
    return _df.loc[:,fb_likes_sum].sum(axis = 'columns')

def cast_like_gt_actor(_df):
    check = _df.cast_total_facebook_likes >= _df.total_likes
    return check

(movies
 .set_index('movie_title')
 .rename(columns=clean_columns)
 .assign(total_likes = sum_likes,
         is_cast_likes_more = cast_like_gt_actor)
 .head()
)
```

### Check to see if the value are all true
is_cast_likes_more is now a column from a Boolean array. We can check whether all the values of this column are True using the .all method:

```python
def clean_columns(val):
    return val.strip().lower().replace(' ','_')

def sum_likes(_df):
    fb_likes_sum = ("actor_1_facebook_likes","actor_2_facebook_likes","actor_3_facebook_likes","director_facebook_likes")
    return _df.loc[:,fb_likes_sum].sum(axis = 'columns')

def cast_like_gt_actor(_df):
    check = _df.cast_total_facebook_likes >= _df.total_likes
    return check

(movies
 .set_index('movie_title')
 .rename(columns=clean_columns)
 .assign(total_likes = sum_likes,
         is_cast_likes_more = cast_like_gt_actor)
 .is_cast_likes_more
 .all()
)
```

### Drop and create another column using list comprehension

It turns out that there is at least one movie with more total_likes than cast_total_facebook_likes. It could be that director Facebook likes are not part of the cast total likes. Let's backtrack and delete the total_likes column. We can use the .drop method with the columns parameter to do that.

Let's recreate a Series of just the total actor likes:

```python
def clean_columns(val):
    return val.strip().lower().replace(' ','_')


def sum_likes(_df):
    fb_likes_sum = ("actor_1_facebook_likes","actor_2_facebook_likes","actor_3_facebook_likes","director_facebook_likes")
    return _df.loc[:,fb_likes_sum].sum(axis = 'columns')


def cast_like_gt_actor(_df):
    check = _df.cast_total_facebook_likes >= _df.actor_sum
    return check

## Use a list comprehension
def actor_likes(_df):
    colNames = [col for col in _df.columns 
            if "actor_" in col and "_likes" in col]

    actor_likes = (_df
                   .loc[:,colNames]
                   .sum(axis="columns"))
    return actor_likes
    

(movies
 .set_index('movie_title')
 .rename(columns=clean_columns)
 .assign(total_likes = sum_likes,
         actor_sum = actor_likes,
         is_cast_likes_more = cast_like_gt_actor)
 .drop(columns = ['total_likes'])
 .head()
)
```

Check again whether all the values in cast_total_facebook_likes are greater than actor_sum. We can do this with the >= operator or the .ge method:

```python
def clean_columns(val):
    return val.strip().lower().replace(' ','_')


def sum_likes(_df):
    fb_likes_sum = ("actor_1_facebook_likes","actor_2_facebook_likes","actor_3_facebook_likes","director_facebook_likes")
    return _df.loc[:,fb_likes_sum].sum(axis = 'columns')


def cast_like_gt_actor(_df):
    check = _df.cast_total_facebook_likes >= _df.actor_sum
    return check

## Use a list comprehension
def actor_likes(_df):
    colNames = [col for col in _df.columns 
            if "actor_" in col and "_likes" in col]

    actor_likes = (_df
                   .loc[:,colNames]
                   .sum(axis="columns"))
    return actor_likes
    

(movies
 .set_index('movie_title')
 .rename(columns=clean_columns)
 .assign(total_likes = sum_likes,
         actor_sum = actor_likes,
         is_cast_likes_more = cast_like_gt_actor)
 .drop(columns = ['total_likes'])
 .is_cast_likes_more
 .all()
)
```

### Calculate using an expression
Finally, let's calculate the percentage of the cast_total_facebook_likes that come from actor_sum:

```python
def clean_columns(val):
    return val.strip().lower().replace(' ','_')


def sum_likes(_df):
    fb_likes_sum = ("actor_1_facebook_likes","actor_2_facebook_likes","actor_3_facebook_likes","director_facebook_likes")
    return _df.loc[:,fb_likes_sum].sum(axis = 'columns')


def cast_like_gt_actor(_df):
    check = _df.cast_total_facebook_likes >= _df.actor_sum
    return check


## Use a list comprehension
def actor_likes(_df):
    colNames = [col for col in _df.columns 
            if "actor_" in col and "_likes" in col]

    actor_likes = (_df
                   .loc[:,colNames]
                   .sum(axis="columns"))
    return actor_likes


def pct_facebook(_df):
    return (_df
            .actor_sum
            .div(_df.cast_total_facebook_likes)
            .mul(100)
           )


def prep_movies(_df):
    df = (_df
         .set_index('movie_title')
         .rename(columns=clean_columns)
         .assign(total_likes = sum_likes,
                 actor_sum = actor_likes,
                 is_cast_likes_more = cast_like_gt_actor,
                 pct_like = pct_facebook
                )
         .drop(columns = ['total_likes'])
        )

    return df
    
prep_movies(movies)
```
