## Selecting columns with methods
Although column selection is usually done with the indexing operator, there are some DataFrame methods that facilitate their selection in an alternative manner. The .select_dtypes and .filter methods are two useful methods to do this.

If you want to select by type, you need to be familiar with pandas data types. The Understanding data types recipe in Chapter 1, Pandas Foundations, explains the types.

### Select columns by data types
Read in the movie dataset. Shorten the column names for display. Use the .get_dtype_counts method to output the number of columns with each specific data type:

```python
def shorten(cols):
    newCols = (str(cols)
               .replace("facebook_likes","fb")
               .replace("_for_reviews","")
    )
    return newCols

(movies
 .rename(columns = shorten)
 .dtypes
 .value_counts())
```

```python
(movies
 .rename(columns = shorten)
 .select_dtypes(include = "int64")
 .head()
)
```

If you would like to select all the numeric columns, you may pass the string number to the include parameter:

```python
(movies
 .rename(columns = shorten)
 .select_dtypes(include = "number")
 .head()
)
```

To exclude only floating-point columns, do the following:

```python
(movies
 .rename(columns = shorten)
 .select_dtypes(exclude = "float64")
 .dtypes
)
```

```python
(movies
 .rename(columns = shorten)
 .select_dtypes(exclude = "float64")
 .head()
)
```

One confusing aspect of .select_dtypes is its flexibility to take both strings and Python objects. The following list should clarify all the possible ways to select the many different column data types. There is no standard or preferred method of referring to data types in pandas, so it's good to be aware of both ways:

- np.number, 'number' – Selects both integers and floats regardless of size
- np.float64, np.float_, float, 'float64', 'float_', 'float' – Selects only 64-bit floats
- np.float16, np.float32, np.float128, 'float16', 'float32', 'float128' – Respectively selects exactly 16, 32, and 128-bit floats
- np.floating, 'floating' – Selects all floats regardless of size
- np.int0, np.int64, np.int_, int, 'int0', 'int64', 'int_', 'int' – Selects only 64-bit integers
- np.int8, np.int16, np.int32, 'int8', 'int16', 'int32' – Respectively selects exactly 8, 16, and 32-bit integers
- np.integer, 'integer' – Selects all integers regardless of size
- 'Int64' – Selects nullable integer; no NumPy equivalent
- np.object, 'object', 'O' – Select all object data types
- np.datetime64, 'datetime64', 'datetime' – All datetimes are 64 bits
- np.timedelta64, 'timedelta64', 'timedelta' – All timedeltas are 64 bits
- pd.Categorical, 'category' – Unique to pandas; no NumPy equivalent
Because all integers and floats default to 64 bits, you may select them by using the string 'int' or 'float' as you can see from the preceding bullet list. If you want to select -all integers and floats regardless of their specific size, use the string 'number'.

### Selecting columns using Filter

The .filter method has more tricks (or parameters) up its sleeve. If you use the items parameters, you can pass in a list of column names:

```python
cols = [
    "actor_1_name",
    "actor_2_name",
    "actor_3_name",
    "director_name",
]
(movies
 .filter(items=cols)
 .head())
```

An alternative method to select columns is with the .filter method. This method is flexible and searches column names (or index labels) based on which parameter is used. Here, we use the like parameter to search for all the 'likes' columns or the names that contain the exact string, likes. The like parameter is checking for substrings in column names:

```python
(movies
 .filter(like = 'likes')
 .head()
)
```

The .filter method allows columns to be searched with regular expressions using the regex parameter. Here, we search for all columns that have a digit somewhere in their name:

```python
(movies
 .filter(regex = r"\d")
 .head()
)
```

```python
movies.head()
```

## Chaining DataFrame methods

We will use the .isnull method to get a count of the missing values. This method will change every value to a Boolean, indicating whether it is missing:

```python
def shorten(col):
    return (col
            .replace('facebook_likes', 'fb')
            .replace('_for_reviews', '')
           )

(movies
 .rename(columns = shorten)
 .isnull()
 .head())
```

We will chain the .sum method that interprets True and False as 1 and 0, respectively. Because this is a reduction method, it aggregates the results into a Series:

```python
(movies
 .rename(columns = shorten)
 .isnull()
 .sum())
```

We can go one step further and take the sum of this Series and return the count of the total number of missing values in the entire DataFrame as a scalar value:

```python
(movies
 .rename(columns = shorten)
 .isnull()
 .sum()
 .sum()
)
```

A way to determine whether there are any missing values in the DataFrame is to use the .any method twice in succession:

```python
(movies
 .isnull()
 .any()
)
```

## Comparing missing values

pandas uses the NumPy NaN (np.nan) object to represent a missing value. This is an unusual object and has interesting mathematical properties. For instance, it is not equal to itself. Even Python's None object evaluates as True when compared to itself:

```python
np.nan == np.nan
```

```python
None == None
```

All other comparisons against np.nan also return False, except not equal to (!=):

```python
np.nan > 5
```

```python
np.nan != 5
```

To get an idea of how the equals operator works, let's compare each element to a scalar value:

```python
college = pd.read_csv(dataPath + '\\college.csv', index_col = 'INSTNM')
college.head()
```

```python
college_ugds = college.filter(like = 'UGDS_')
college_ugds.head()
```

To get an idea of how the equals operator works, let's compare each element to a scalar value:

```python
college_ugds == 0.0019
```

This works as expected but becomes problematic whenever you attempt to compare DataFrames with missing values. You may be tempted to use the equals operator to compare two DataFrames with one another on an element-by-element basis. Take, for instance, college_ugds compared against itself, as follows:

```python
college_self_compare = college_ugds == college_ugds
college_self_compare.head()
```

At first glance, all the values appear to be equal, as you would expect. However, using the .all method to determine if each column contains only True values yields an unexpected result:

```python
college_self_compare.all()
```

This happens because missing values do not compare equally with one another. If you tried to count missing values using the equal operator and summing up the Boolean columns, you would get zero for each one:

```python
(college_ugds == np.nan).sum()
```

Instead of using == to find missing numbers, use the .isna method:

```python
(college_ugds
 .isna()
 .sum()
)
```

The correct way to compare two entire DataFrames with one another is not with the equals operator (==) but with the .equals method. This method treats NaNs that are in the same location as equal (note that the .eq method is the equivalent of ==):

```python
college_ugds.equals(college_ugds)
```

## Transposing the direction of a DataFrame operation

```python
college_ugds.head()
```

Now that the DataFrame contains homogenous column data, operations can be sensibly done both vertically and horizontally. The .count method returns the number of non-missing values. By default, its axis parameter is set to 0:

```python
college_ugds.count()
```

The axis parameter is almost always set to 0. So, step 2 is equivalent to both college_ugds.count(axis=0) and college_ugds.count(axis='index').

Changing the axis parameter to 'columns' changes the direction of the operation so that we get back a count of non-missing items in each row:

```python
college_ugds.count(axis="columns").head(10)
```

Instead of counting non-missing values, we can sum all the values in each row. Each row of percentages should add up to 1. The .sum method may be used to verify this:

```python
college_ugds.sum(axis = 'columns')
```

The .cumsum method with axis=1 accumulates the race percentages across each row. It gives a slightly different view of the data. For example, it is very easy to see the exact percentage of white and black students for each school:

```python
college_ugds.cumsum(axis = 'columns').head()
```

## Determining college campus diversity

Many articles are written every year on the different aspects and impacts of diversity on college campuses. Various organizations have developed metrics attempting to measure diversity. US News is a leader in providing rankings for many different categories of colleges, with diversity being one of them. Their top 10 diverse colleges with Diversity Index are given as follows:

```python
college = pd.read_csv(dataPath + '\\college.csv', index_col = 'INSTNM')
college.head()
```

```python
college_ugds = college.filter(like = 'UGDS_')
college_ugds.head()
```

```python
college_ugds.shape
```

```python
college_ugds.sum(axis = 'columns')
```

```python
college_ugds.ge(.15)
```

Many of these colleges have missing values for all their race columns. We can count all the missing values for each row and sort the resulting Series from the highest to lowest. This will reveal the colleges that have missing values:

```python
(college_ugds
 .isnull()
 .sum(axis = 'columns')
 .sort_values(ascending = False)
)
```

Now that we have seen the colleges that are missing all their race columns, we can use the .dropna method to drop all rows that have all nine race percentages missing. We can then count the remaining missing values:

```python
college_ugds = (
  college_ugds
 .dropna(axis = 'index', how = 'any')
)
```

```python
(college_ugds
 .isnull()
 .sum()
)
```

There are no missing values left in the dataset. We can now calculate our diversity metric. To get started, we will use the greater than or equal DataFrame method, .ge, to return a DataFrame with a Boolean value for each cell:

```python
college_ugds.ge(.15)
```

From here, we can use the .sum method to count the True values for each college. Notice that a Series is returned:

```python
college_ugds.head()
```

```python
(college_ugds
 .ge(.15)
 .sum())
```

```python
diversity_metric = (
 college_ugds
 .ge(.15)
 .sum(axis = 'columns')
)

diversity_metric
```

To get an idea of the distribution, we will use the .value_counts method on this Series:

```python
diversity_metric.value_counts()
```

```python
diversity_metric.value_counts(normalize = True)
```

Amazingly, two schools have more than 15% in five different race categories. Let's sort the diversity_metric Series to find out which ones they are:

```python
diversity_metric.sort_values(ascending = False).head()
```

It seems a little suspicious that schools can be that diverse. Let's look at the raw percentages from these top two schools. We will use .loc to select rows based on the index label:

```python
veryDiverseSchools_gt5 = (
 diversity_metric
 .sort_values(ascending = False)
 .head(2)
 .index
 .to_list()
)

college_ugds.loc[veryDiverseSchools_gt5]
```

It appears that several categories were aggregated into the unknown and two or more races column. Regardless of this, they both appear to be quite diverse. We can see how the top five US News schools fared with this basic diversity metric:

```python
us_news_top = [
    "Rutgers University-Newark",
    "Andrews University",
    "Stanford University",
    "University of Houston",
    "University of Nevada-Las Vegas",
]

diversity_metric.loc[us_news_top]
```

Alternatively, we can find the schools that are least diverse by ordering them by their maximum race percentage:

```python
(college_ugds
 .max(axis = 'columns')
 .sort_values(ascending = False)
 .head(10)
)
```

We can also determine if any school has all nine race categories exceeding 1%:

```python
college_ugds.gt(.01).all(axis = 'columns').any()
```

```python
x = college_ugds.gt(.01).sum(axis = 'columns')
```

```python
x[x == 9].count()
```
