```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import os

path = os.getcwd() + '\\data\\'
file = 'AAPL.csv'
```

```python
df_raw = pd.read_csv(path + file)
df_raw.head()
```

Alternate method when reading data: df_alt = pd.read_csv(path + file, parse_dates=['Date'])

```python
df_raw.info(verbose=True)
```

```python
def prep_data(_df):
    return (_df
            .astype({'Date': 'datetime64[ns]'})
            .set_index('Date')
            .assign(Year = lambda df:df.index.year,
                    Month = lambda df:df.index.month)
    )
    
df = prep_data(df_raw)
df.head()
```

## Basic data manipulations

Apply a function along a Series

```python
df.Open.map(lambda col: col + 100)
```

```python
def add100(_column):
    return _column + 100

df.Open.map(add100)
```

## Group aggregations

```python
df.head()
```

```python
df.loc['1980',['Volume']].sum()
```

```python
(df
 .loc[:,['Volume','Year']]
 .groupby('Year')
 .sum()
)
```

## Pivot table

```python
df.loc[:'2010']
```

```python
pd.pivot_table(df.loc['2010':'2020'], 
               index='Month', 
               columns = 'Year', 
               values = 'Open', 
               aggfunc = np.mean)
```

```python
pd.pivot_table(df.loc['2010':'2020'], 
               index='Month', 
               columns = 'Year', 
               values = 'Open', 
               aggfunc = np.mean).plot(figsize=(14,10))
```

```python
df.plot.scatter(x='Open',y='Close', figsize = (14,10))
```

```python
(df
 .loc[:,['Year','Open']]
 .groupby('Year')
 .mean()
)
```

```python
(df
 .loc[:,['Year','Open']]
 .groupby('Year')
 .mean()
 .plot(figsize = (14,10))
)
```
