# Simple Time Series Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (12, 8)
```

## GDP 

```python
gdp = pd.read_csv('data/GDP.csv', 
                  parse_dates=['DATE'], 
                  index_col = 'DATE' )
```

```python
gdp.head()
```

```python
gdp.info(verbose = True)
```

```python
ax = gdp.plot(legend = False)
ax.set_ylabel('GDP (Billions)');
```

## Influenza 

```python
ill = pd.read_csv('data/CDC.csv')
ill.head()
```

```python
ill['Date'] = ill.Year + (ill.Week/52)
```

```python
ill.head()
```

```python
colNames = ['Percent of Deaths Due to Pneumonia and Influenza', 'Expected', 'Threshold']

ax = ill.plot(x = 'Date', y = colNames)
ax.legend(['Mortality','Expected','Threshold'])
ax.set_xlabel('')
ax.set_ylabel('% Mortality')
ax.set_ylim(bottom=0, top = 12);
```

## Sun

```python
sun = pd.read_csv('data/sun.csv', dtype={'YEAR':'Int32'}, index_col = 'YEAR')
sun.head()
```

```python
ax = sun.plot(legend=False)
ax.set_ylabel('Sun Spot Activity')
ax.set_xlabel('')
```

## Dow-Jones Industrial Average

```python
djia = pd.read_csv('data/DJIA.csv', parse_dates=['DATE'], index_col='DATE', na_values = '.')
djia.head()
```
