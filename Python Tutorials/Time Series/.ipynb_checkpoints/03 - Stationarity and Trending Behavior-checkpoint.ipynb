# Stationarity and Trending Behavior

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (12, 8)
```

## Create Data

```python
x = np.linspace(0, np.pi*10, 360)
y = np.sin(x)

df = pd.DataFrame({'x':x, 'y':y})
```

```python
df.head()
```

## Non-stationarity

```python
fix, axes = plt.subplots(2,2, figsize=(16,12))

axes[0][0].plot(x, y)
axes[0][0].set_title('Stationary Series')

axes[0][1].plot(x, y+(x/10))
axes[0][1].set_title('Changing Mean')

axes[1][0].plot(x, y*(x/10))
axes[1][0].set_title('Changing Variance')

axes[1][1].plot(np.sin(x+x*x/30))
axes[1][1].set_title('Changing Co-variance')

plt.tight_layout()
```

## Trend 
Two of the most common non-stationarity types are due to different kinds of trends that we illustrate here:

```python
fig, axes = plt.subplots(1,2, figsize = (18,9))

axes[0].plot(x, y+(x/10))
axes[0].set_title('Addititve Trend')

axes[1].plot(x, y*(x/10))
axes[1].set_title('Multiplicative Trend')
```

## Empirical Time Series
We also saw several examples of non-stationary time series. A good example is the international airline passengers dataset.

```python
air = pd.read_csv('data/international-airline-passengers.csv', 
                  sep=';', 
                  parse_dates=['Month'],
                  index_col='Month')
air.head()
```

```python
fig, ax = plt.subplots()
ax.plot(air)
ax.set_ylabel('Number of Passengers')
```

One quick way to view a trend is to look at the average number of passengers per year.

```python
air.head()
```

```python
air_sum = (air
           .assign(Year = air.index.year)
           .groupby('Year')
           .agg({'Passengers':['mean','max','min']})
           .round()
           .reset_index()
           .assign(EndOfYearAverage = lambda _df: pd.to_datetime(_df.Year.astype(str)+'-12-31'),
                   Range = lambda _df: _df.Passengers['max'] - _df.Passengers['min'])
)

air_sum.columns = ['Year','MeanPassengers', 'MaxPassengers','MinPassengers', 'EndOfYearAverage','Range']
air_sum
```

```python
fig, ax = plt.subplots()

ax.plot(air, color='#BDBDBD', label='Monthly Total')
ax.plot(air_sum.EndOfYearAverage, air_sum.MeanPassengers, label='Yearly Average')
ax.set_title('Number of Passengers in the Air')
ax.set_ylabel('Number of Passengers')

ax.text(air.index[-1], air.iloc[len(air)-1], 'Monthly Total', fontdict={'color':'gray'})
ax.text(air_sum['EndOfYearAverage'].iloc[len(air_sum)-1], air_sum['MeanPassengers'].iloc[len(air_sum)-1], 'Yearly Average', fontdict={'color':'#167DB6'});
```

```python
fig, ax = plt.subplots()
ax.errorbar(air_sum.EndOfYearAverage,air_sum.MeanPassengers, 
            yerr=air_sum.Range, ecolor='skyblue')
ax.set_ylabel('Number of Passengers')
```
