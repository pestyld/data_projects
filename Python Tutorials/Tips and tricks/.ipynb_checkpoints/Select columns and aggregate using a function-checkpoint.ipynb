```python
import pandas as pd
```

```python
tbl = {'my column':[1,2,3,4,5], 
       'MY_OTHER col':['a','ab','b','c','d'],
       'abc1':[1,2,3,4,5],
       'abc2':[1,2,3,4,5],
       'abc3':[1,2,3,4,5]}

df = pd.DataFrame(tbl)

df.head()
```

```python
def sum_likes(_df):
    sum_cols = []
    for column in _df.columns:
        if "a" in column and ("b" in column or "c" in column):
            sum_cols.append(column)
            
    return _df.loc[:,sum_cols].sum(axis = "columns")



(df
 .assign(total_likes = sum_likes)
 .head()
)
```
