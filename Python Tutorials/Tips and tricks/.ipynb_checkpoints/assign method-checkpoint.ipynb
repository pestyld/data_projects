## Use the assign method
Convert the MSRP and Invoice columns to numeric. Use the original dataframe column values.

```python
(df
 .assign(
         MSRP = df.MSRP.str.replace('$','', regex=True).str.replace(',','', regex=True).astype('int'),
         Invoice = df.Invoice.str.replace('$','', regex=True).str.replace(',','', regex=True).astype('int')
        )
)
```

## Use a lambda function within the assign method
The lambda function uses the intermediate dataframe

```python
(df
 .assign(
         MSRP = lambda _df: _df.MSRP.str.replace('$','', regex=True).str.replace(',','', regex=True).astype('int'),
         Invoice = lambda _df:_df.Invoice.str.replace('$','', regex=True).str.replace(',','', regex=True).astype('int')
        )
)
```

## Use a user function within the assign method

```python
def makeNumber(_df, colName):
    return (_df[colName]
            .str.replace('$','', regex=True)
            .str.replace(',','', regex=True)
            .astype('int')
    )

(df
 .assign(
         MSRP = makeNumber(df,'MSRP'),
         Invoice = makeNumber(df,'Invoice')
        )
)
```

## Use a user function in the assign method within a lambda function

```python
def makeNumber(_df, colName):
    return (_df[colName]
            .str.replace('$','', regex=True)
            .str.replace(',','', regex=True)
            .astype('int')
    )

(df
 .assign(
         MSRP = lambda _df:  makeNumber(_df,'MSRP'),
         Invoice = lambda _df: makeNumber(_df,'Invoice')
        )
)
```

## Use the intermdiate dataframe in assign to calculate values

Try and subtract the new numeric Invoice and MSRP columns.

1. Start by checking with value will be used since the original columns are overwritten. Notice using the df.MSRP using the ORIGINAL MSRP column.

```python
def makeNumber(_df, colName):
    return (_df[colName]
            .str.replace('$','', regex=True)
            .str.replace(',','', regex=True)
            .astype('int')
    )

(df
 .assign(
         MSRP = makeNumber(df,'MSRP'),
         Invoice = makeNumber(df,'Invoice'),
         PriceDiff = df.MSRP
        )
)
```

2. Using a lambda function in the assign method pulls the intermediate dataframe value. In this case the numeric value in the MSRP column.

```python
def makeNumber(_df, colName):
    return (_df[colName]
            .str.replace('$','', regex=True)
            .str.replace(',','', regex=True)
            .astype('int')
    )

(df
 .assign(
         MSRP = makeNumber(df,'MSRP'),
         Invoice = makeNumber(df,'Invoice'),
         PriceDiff = lambda _df: _df.MSRP
        )
)
```

3. Calculate the price difference between MSRP and Invoice.

```python
def makeNumber(_df, colName):
    return (_df[colName]
            .str.replace('$','', regex=True)
            .str.replace(',','', regex=True)
            .astype('int')
    )

(df
 .assign(
         MSRP = makeNumber(df,'MSRP'),
         Invoice = makeNumber(df,'Invoice'),
         PriceDiff = lambda _df: _df.MSRP - _df.Invoice
        )
)
```

4. Use a function to calcuate the difference

```python
def makeNumber(_df, colName):
    return (_df[colName]
            .str.replace('$','', regex=True)
            .str.replace(',','', regex=True)
            .astype('int')
    )

def priceDiff(_df):
    return _df.MSRP - _df.Invoice

(df
 .assign(
         MSRP = makeNumber(df,'MSRP'),
         Invoice = makeNumber(df,'Invoice'),
         PriceDiff = lambda _df: priceDiff(_df)
        )
)
```

```python
4. PriceDiff percentage
```

```python
def makeNumber(_df, colName):
    return (_df[colName]
            .str.replace('$','', regex=True)
            .str.replace(',','', regex=True)
            .astype('int')
    )

def priceDiffPct(_df):
    return (((_df.MSRP - _df.Invoice)/_df.MSRP)
            .mul(100)
            .round(1)
    )

(df
 .assign(
         MSRP = makeNumber(df,'MSRP'),
         Invoice = makeNumber(df,'Invoice'),
         PriceDiff = lambda _df: priceDiffPct(_df)
        )
)
```

## Creating a boolean value using assign

```python
def makeNumber(_df, colName):
    return (_df[colName]
            .str.replace('$','', regex=True)
            .str.replace(',','', regex=True)
            .astype('int')
    )

(df
 .assign(
         MSRP = makeNumber(df,'MSRP'),
         Invoice = makeNumber(df,'Invoice'),
         PriceDiff = lambda _df: _df.MSRP - _df.Invoice,
         MSRPGreaterThanInvoice = lambda _df: _df.MSRP >= _df.Invoice
        )
)
```
