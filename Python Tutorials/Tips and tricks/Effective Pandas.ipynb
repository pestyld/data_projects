```python
def tweak_autos(autos):
    cols = ['city08', 'comb08', 'highway08', 'cylinders', 'displ', 'drive','eng_dscr', 'fuelCost08', 'make',
            'model', 'trany', 'range', 'createdOn', 'year']
    
    modify_dtypes = {'highway08':'int8', 'city08':'int16', 'comb08':'int16', 
                 'fuelCost08':'int16', 'range':'int16', 'year':'int16',
                 'make':'category'}

    return (autos[cols]
    ## Display the dataframe at this point in time
    .pipe(lambda _df: display(_df) or _df) 
            
    ## Create new calculated columns
    .assign(
         cylinders = autos.cylinders.fillna(0).astype('int8'),
         displ = autos.displ.fillna(0).astype('float16'),
         drive = autos.drive.fillna('Other').astype('category'),
         automatic = autos.trany.str.contains('Auto'),
         speeds = autos.trany.str.extract(r'(\d)+').fillna('20').astype('int8'),
         createdOn = pd.to_datetime(autos.createdOn.replace({' EDT':'-04:00', ' EST':'-05:00'}, regex = True)),
         FFS = autos.eng_dscr.str.contains('FFS')
        )
    
    ## Change the data types using the varaible above
    .astype(modify_dtypes)
            
    ## Drop unncessary columns
    .drop(columns = ['trany', 'eng_dscr'])
    )

autos2 = tweak_autos(autos)
```

### Apply a user function to the agg method

```python
def second_to_last(ser):
    return ser.iloc[-2]

(autos2
 .assign(country = autos2.make.apply(country))
 .groupby(['year', 'country'])
 .agg(['min', 'mean', second_to_last])
)
```

### Unstack

```python
(autos2
 .assign(country = autos2.make.apply(country))
 .groupby(['year', 'country'])
 .mean()
 .unstack()
)
```

### Simple plot for multi index

```python
(autos2
 .assign(country = autos2.make.apply(country))
 .groupby(['year', 'country'])
 .mean()
 .unstack()
 .city08
 .plot(figsize = (10,6))
 .legend(loc = 'upper left')
)
```

### Smoothing

```python
(autos2
 .assign(country = autos2.make.apply(country))
 .groupby(['year', 'country'])
 .mean()
 .unstack()
 .city08
 .rolling(2)
 .mean()
 .plot(figsize = (10,6))
 .legend(loc = 'upper left')
);
```

### Filtering groups like you would SQL group by with having

#### Using Pipe

```python
def vals_gt(_df, num):
    return _df[_df.gt(num)].dropna()


(autos2
 .assign(country = autos2.make.apply(country))
 .groupby(['year', 'country'])
 .count()
 .pipe(vals_gt, 700)
)
```
