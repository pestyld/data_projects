# Merge and Join DataFrames

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'leftkey':['foo','bar','bz','foo'], 'value':[1,2,3,4]})
df2 = pd.DataFrame({'rightkey':['foo','bar','quz','bar'], 'value':[5,6,7,8]})
display(df1,df2)
```

## Outer

```python
df1.merge(df2, left_on = 'leftkey', right_on = 'rightkey', how = 'outer')
```

## Inner

```python
df1.merge(df2, left_on = 'leftkey', right_on = 'rightkey', how = 'inner')
```

## Join Using Index

```python
display(df1, df2)
```

```python
df1.set_index('leftkey', inplace = True)
df2.set_index('rightkey', inplace = True)
display(df1, df2)
```

```python
df1.join(df2, lsuffix='_l', rsuffix='_r', how = "inner")
```

```python
df1.join(df2, lsuffix='_l', rsuffix='_r', how = "outer")
```
