## Prepare the Data

```python
df = (
    df_raw
    .rename(columns = lambda colname: colname.replace('.',' ').replace(' ',''))
)

df.head()
```

## Creating training and test data

```python
## Target
y = df['Price']

## Features
X = df[['AvgAreaIncome', 'AvgAreaHouseAge', 'AvgAreaNumberofRooms',
       'AvgAreaNumberofBedrooms', 'AreaPopulation']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
## Create linear regression object
lm = LinearRegression()

type(lm)
```

```python
## Fit the model
lm.fit(X_train,y_train)
```

```python
## print intercept
lm.intercept_
```

```python
## coefficients for each column in X
lm.coef_
```

```python
pd.DataFrame(lm.coef_,          ## First column
             X_train.columns,   ## Index
             columns = ['Coeff'])
```

All other features fixed:
- A one unit increase in **AvgAreaIncome** is associated with a \$21 increase in price.
- A one unit increase in **AvgAreaHouseAge** is associated with a \$164,883 increase in price.
- A one unit increase in **AvgAreaNumberofRooms** is associated with a \$122,368. increase in price.
- A one unit increase in **AvgAreaNumberofBedrooms** is associated with a \$2,233 increase in price.
- A one unit increase in **AreaPopulation** is associated with a \$15 increase in price.

## Predication from the model

```python
## Add the features of the test dataframe to predict the housing prices
predictions = lm.predict(X_test)

display(type(predictions), predictions)
```

### Compare predicted vs real on test data

```python
## Compare the actual vs predicted
fig, ax = plt.subplots()
ax.scatter(x = predictions, y = y_test, edgecolors='black', s=20)
# ax.yaxis.set_major_formatter('${x:1,.2f}')
# ax.xaxis.set_major_formatter('${x:1,.2f}')

plt.ticklabel_format(style='plain', axis='x')
plt.ticklabel_format(style='plain', axis='y');
```

## Regression Evaulation Metrics

- **Mean absolute error (MAE)** - Mean of the absolute value of the errors.
- **Mean squared error (MSE)** - Mean of the squared errors
- **Root mean squared error (RMSE)** - Square root of the mean of the squared errors.

All ways to compare prediction vs true values.

Comparing metrics:
- **MAE** : Easiest to understand because it's the average error.
- **MSE** : More popular than MAE because MSE punishes larger errors, which tends to be useful in the real world.
- **RMSE** - More popular than MSE because RMSE is interpretable in the *y* units.

All of these are **loss functoins**, because we want to minimize these *errors*. To create the best model.

```python
MAE = metrics.mean_absolute_error(y_test, predictions)
MSE = metrics.mean_squared_error(y_test, predictions)
RMSE = np.sqrt(MSE)

print('MAE:', MAE)
print('MSE:', MSE)
print('RMSE:',RMSE)
```
